# INSTRUCTION CONDITIONNELLE : IF .... ELIF ... ELSE 


## Objectif : effectuer des actions seulement si une  certaine condition est vérifiée 


### Syntaxe en pyhon : if ...


<code>
if condition :
    instructions à excecuter si vrai
</code>
    
> la condition est une expression booléenne.

> Attention à l'indentation ! 
- indique dans quel bloc se trouve une instruction
- **obligatoire** en python
    


## Objectif : effectuer des actions différentes selon qu'une certaine condition est vérifiée ou pas


### Syntaxe en pyhon : if ... else 

<code>
if condition :   
  instructions à exécuter si vrai 
else : 
  instructions à exécuter si faux 
</code>
    
    
> Attention : le <code>else</code> n’est pas suivi d’une condition !

```python
### Exemple :

x = float(input("Entrez un nombre :"))   
if x > 0 : 
 print(x, "est plus grand que 0")   
 print("il est strictement positif") 
else : 
 print(x, "est négatif ou nul") 
print("Fin")
```


## Objectif : enchaîner plusieurs conditions 


### Syntaxe en pyhon : if ... elif ... else ...

<code>
if condition1 :   
  instructions à exécuter si condition1 vrai 
elif condition2 : 
  #condition1 faux
  instructions à exécuter si condition2 vrai 
else :   
  instructions à exécuter si condition1 et condition 2 fausse
</code>

```python
### Exemple :

x = float(input("Entrez un nombre :"))   
if x > 0 : 
 print(x, "est plus grand que 0")   
 print("il est strictement positif") 
elif (x<0) : 
 print(x, "est négatif")
else :
 print(x, "vaut 0")

print("Fin")
```
